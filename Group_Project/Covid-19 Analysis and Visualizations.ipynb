### First Objective: Determine how the three datasets compare. Are there any consistencies in data? Is there any bias in the data sources?

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Load the csv data files
data_source_1 = 'output_data/data_source_1.csv'
data_source_2 = 'output_data/data_source_2.csv'
data_source_3 = 'output_data/data_source_3.csv'

# Read the Covid-19 data and the study results
data_1 = pd.read_csv(data_source_1)
data_2 = pd.read_csv(data_source_2)
data_3 = pd.read_csv(data_source_3)

# Change the date to data/time format for plotting
data_1['date'] = pd.to_datetime(data_1['date'])
data_2['date'] = pd.to_datetime(data_2['date'])
data_3['date'] = pd.to_datetime(data_3['date'])

print(data_1.dtypes)
```

```python
# Preview first data source
data_1
```

```python
# Generate a line plot of date versus positive cases for each of the 5 different states 
florida1 = data_1.loc[data_1['state'] == 'Florida']
georgia1 = data_1.loc[data_1['state'] == 'Georgia']
newyork1 = data_1.loc[data_1['state'] == 'New York']
texas1 = data_1.loc[data_1['state'] == 'Texas']
washington1 = data_1.loc[data_1['state'] == 'Washington']

# Plot each state separately on same graph
plt.plot(florida1['date'], florida1['positive cases'], color='green', label='Florida')
plt.plot(georgia1['date'], georgia1['positive cases'], color='blue', label='Georgia')
plt.plot(newyork1['date'], newyork1['positive cases'], color='blueviolet', label='New York')
plt.plot(texas1['date'], texas1['positive cases'], color='orange', label='Texas')
plt.plot(washington1['date'], washington1['positive cases'], color='red', label='Washington')

# Define labels of plot
plt.title('Covid-19 Cases for 5 Select States')
plt.xlabel('Date')
plt.ylabel('Total Positive Cases (Per Day)')
plt.xticks(rotation=45)
plt.legend(loc="best")
plt.show()
```
